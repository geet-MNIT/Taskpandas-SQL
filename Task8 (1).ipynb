```python
###5 . which id is one of the active id that you have in whole dataset
mydb = conn.connect(host="localhost",user="root", database="tasksql", passwd="mysql",use_pure=True)
cursor = mydb.cursor()
query5="select identity, count(identity) from tasksql.table1 group by identity order by count(identity) DESC limit 1;"
cursor.execute(query5)
for i in cursor.fetchall():
    print(i)
```

```python
#37 . Find out who is the laziest person id that we have in dataset 
cursor = mydb.cursor()
query7="select distinct identity, sum(TotalSteps) from table1 group by TotalSteps order by TotalSteps limit 1;"
cursor.execute(query7)
for i in cursor.fetchall():
    print(i)
```

```python
##8 .Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 
cursor = mydb.cursor()
query7="select count(identity) from (select identity, sum(Calories) from table1 group by identity having sum(Calories) >30000) table1;"
cursor.execute(query7)
for i in cursor.fetchall():
    print(i)
```

```python
##9. how many person are not a regular person with respect to activity try to find out those
mydb = conn.connect(host="localhost",user="root", database="tasksql", passwd="mysql",use_pure=True)
cursor = mydb.cursor()
query7="select count(*) from (select count(identity) from table1 group by identity having count(identity)!=32) table1;"
cursor.execute(query7)
for i in cursor.fetchall():
    print(i)
```

```python
##10. who is the third most active person in this dataset find out those in pandas and in sql both.
mydb = conn.connect(host="localhost",user="root", database="tasksql", passwd="mysql",use_pure=True)
cursor = mydb.cursor()
query7="select identity, count(identity) from table1 group by identity order by count(identity) desc, identity desc limit 3,1;"
cursor.execute(query7)
for i in cursor.fetchall():
    print(i)
```
